```python
###  Reset Index , set index ,indexing and slicing of index object get_loc , is_unique , create index object from scratch
```

```python
import pandas as pd
```

```python
#  set column as an index
movie=pd.read_csv("tmdb_5000_movies.csv" ,index_col="original_title")
movie
```

```python
#  reset index
movie.reset_index()
```

```python
movie
#  here index not reset
```

```python
movie.reset_index(inplace=True)
#  here now index is reset in original dataframe
movie
```

```python
#  replace value of index
movie.set_index("id",inplace=True)
movie
```
