```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")

tips_df
```

```python
#plot seaborn pairplot
sns.pairplot(tips_df)
```

```python
# vars parameter
sns.pairplot(tips_df, vars = ['total_bill', 'tip', ])
```

```python
# hue and hue_order parameter
sns.pairplot(tips_df, vars = ['total_bill', 'tip', ], hue ='sex')
```

```python
#hue_order parameter
sns.pairplot(tips_df, vars = ['total_bill', 'tip', ], hue ='sex', hue_order=["Female","Male"])
```

```python
# palette parameter
sns.pairplot(tips_df, vars = ['total_bill', 'tip', ], hue ='sex', hue_order=["Female","Male"], palette='Dark2')
```

```python
# x_vars , y_vars parameter
sns.pairplot(tips_df,  hue ='sex', x_vars = ["total_bill", 'tip'], y_vars =['size'])
```

```python
#kind parameter
sns.pairplot(tips_df,  hue ='sex', x_vars = ["total_bill", 'tip'], y_vars =['size'] , kind="reg")
```

```python
# diag_kind parameter
sns.pairplot(tips_df, vars = ['total_bill', 'tip', ], hue ='sex', hue_order=["Female","Male"], palette='Dark2' , diag_kind="hist")
```

```python
# markers parameter
sns.pairplot(tips_df, vars = ['total_bill', 'tip', ], hue ='sex', hue_order=["Female","Male"],  markers = ['*', "<"] , diag_kind="hist")
```

```python
# height parameter
sns.pairplot(tips_df,  hue ='sex', x_vars = ["total_bill", 'tip'], y_vars =['size'] , kind="reg" , height=20)
```
