```python
import seaborn as sns
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

```python
sns.relplot("age" , "fare" , data = titanic_df)
```

```python
sns.relplot("age" , "fare" , data = titanic_df , col = "pclass")
```

```python
sns.relplot("age" , "fare" , data = titanic_df , col = "pclass" , row="sibsp")
```

```python
sns.relplot("age" , "fare" , data = titanic_df , col = "pclass" ,hue="who")
```
