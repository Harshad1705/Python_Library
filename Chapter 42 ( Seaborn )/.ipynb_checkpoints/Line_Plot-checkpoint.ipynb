# How to Draw Seaborn Line Plot ?

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
number = [1,2,3,4,5]
mul = [5,10,15,20,25]

mul_df = pd.DataFrame( { "Number" : number , "Multiplication" : mul})

# lineplot -- takes data in form of dataframe only 
# data parameter of lineplot decribe dataframe 

sns.lineplot( x="Number" , y="Multiplication" , data=mul_df)

plt.show()  # to hide object data
```

```python
# load_datset -- use for load data from github and takes only name of csv file 

tip = sns.load_dataset("tips")
tip
```

```python
sns.lineplot( x="total_bill" , y="tip" , data = tip)
```

```python
sns.lineplot( x="size" , y="total_bill" , data = tip , hue = "sex" , style= "sex" , palette="hot" , dashes=False , markers=["o",">"] , legend=False)
```

```python

# some parameter used with lineplot
sns.lineplot( x="size" , y="total_bill" , data = tip , hue = "sex" , style= "sex" , palette="hot" , dashes=False , markers=["o",">"] , legend=False)

# change title and label
plt.title("Line Plot" , fontsize=20)
plt.xlabel("Sizet" , fontsize=15)
plt.ylabel("Total Bill" , fontsize=15)
```

```python
# change background color
sns.set(style="dark")

# change size of figure
plt.figure(figsize=(16,9))

# some parameter used with lineplot
sns.lineplot( x="size" , y="total_bill" , data = tip , hue = "sex" , style= "sex" , palette="hot" , dashes=False , markers=["o",">"] , legend=False)

# change title and label
plt.title("Line Plot" , fontsize=20)
plt.xlabel("Sizet" , fontsize=15)
plt.ylabel("Total Bill" , fontsize=15)

```
